```python
def create_instruction(row):
    instruction_template = """Instruction:\n{}\n\nResponse:\n{}""".format(row['question'], row['answer'])
    return instruction_template
```

```python
df['ro'] = df.apply(create_instruction, axis=1)
df
```

```python
import keras
import keras_nlp
```

```python
data_list = df['ro'].tolist()
print(data_list[0])
print(len(data_list[0]))
```

```python
gemma_lm = keras_nlp.models.GemmaCausalLM.from_preset("gemma2_2b_en")
gemma_lm.summary()
```

```python
gemma_lm.backbone.enable_lora(rank=4)
gemma_lm.summary()
```

```python
# Limit the input sequence length to 256 (to control memory usage).
gemma_lm.preprocessor.sequence_length = 256
# Use AdamW (a common optimizer for transformer models).
optimizer = keras.optimizers.AdamW(
    learning_rate=5e-5,
    weight_decay=0.01,
)
```

```python
optimizer.exclude_from_weight_decay(var_names=["bias", "scale"])
```

```python
gemma_lm.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=optimizer,
    weighted_metrics=[keras.metrics.SparseCategoricalAccuracy()],
)
```

```python
gemma_lm.fit(data_list, epochs=10, batch_size=1)
```

```python
prompt = """Instruction:\nwhat are the limitations of Generative AI??\n\nResponse:\n"""
```

```python
sampler = keras_nlp.samplers.TopKSampler(k=5, seed=2)
gemma_lm.compile(sampler=sampler)
```

```python
print(gemma_lm.generate(prompt, max_length=2500))
```

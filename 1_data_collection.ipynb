```python
from PyPDF2 import PdfReader
```

```python
reader = PdfReader('/content/interview_3.pdf')
```

```python
num_pages = len(reader.pages)
pdf_text = ""
for page_num in range(num_pages):
    page = reader.pages[page_num]
    pdf_text += page.extract_text()
```

```python
import re

# Pattern to match questions (e.g., "1. What is Generative AI?")
question_pattern = re.compile(r"^\d+\.\s(.*?)\n", re.MULTILINE)

# Pattern to match answers (from question to the next question or end of text)
# Changed answer_pattern to avoid variable-width look-behind
answer_pattern = re.compile(r"(\d+\.\s.*?\n)(.*?)(?=\d+\.\s|\Z)", re.DOTALL | re.MULTILINE)

# Extract questions and answers
questions = question_pattern.findall(pdf_text)
answers_with_questions = answer_pattern.findall(pdf_text)  # Get answers with preceding questions
answers = [answer for _, answer in answers_with_questions]  # Extract only answers

# Store or process the extracted data
qa_pairs = list(zip(questions, answers))

for question, answer in qa_pairs:
    print("Question:", question)
    print("Answer:", answer.strip())  # Strip leading/trailing whitespace from answers
    print("---")
```

```python
import pandas as pd
data = pd.DataFrame(qa_pairs, columns=['question', 'answer'])
```

```python
data['answer'] = data['answer'].str.strip()
```

```python
reader1 = PdfReader('/content/interview_2.pdf')
```

```python
num_pages = len(reader1.pages)
pdf_text_1 = []
for page_num in range(num_pages):
    page = reader1.pages[page_num]
    pdf_text_1.append([page_num, page.extract_text()])
```

```python
text = ''
for i in pdf_text_1[94:164]:
    text += i[1]
print(text)
```

```python
question_pattern = re.compile(r"Q\d+\)\s(.*?)\n", re.MULTILINE)  # Modified for "Q" format

# Pattern to match answers (until the next question or end of text)
answer_pattern = re.compile(r"(Q\d+\)\s.*?\n)(.*?)(?=Q\d+\)\s|\Z)", re.DOTALL | re.MULTILINE)  # Modified

# Extract questions and answers
questions = question_pattern.findall(text)
answers_with_questions = answer_pattern.findall(text)
answers = [answer for _, answer in answers_with_questions]
```

```python
# Create a Pandas DataFrame
data2 = pd.DataFrame({'question': questions, 'answer': answers})
```

```python
data = pd.concat([data, data2], ignore_index=True)
```

```python
data.to_csv('qa_data.csv', index=False)  # 'index=False' prevents writing row indices to the file
```
